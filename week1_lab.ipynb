```python
scores = [72 , 85 , 90 , 68 , 88]
print("Scores:", scores)

average = sum(scores) / len(scores)
print("Average score:", average)

high_score = max(scores)
lowest_score = min(scores)

print("Max and min values:", high_score, lowest_score)
```

```python
study_hours = [1 , 2 , 3 , 4 , 5]
exam_scores = [50 , 55 , 65 , 70 , 80]

for hours, marks in zip(study_hours, exam_scores):
    print(f"Input: {hours} hours → Output: {marks} marks")
```

```python
scores = [45 , 67 , 82 , 59]

def pass_fail(score: int):
  if score >= 60:
    return "Pass"
  else:
    return "Fail"

for s in scores:
  print(pass_fail(s))
```

```python
sizes = [500 , 800 , 1000 , 1200]
prices = [50 , 80 , 100 , 130]

for size, price in zip(sizes, prices):
    print(f"Size: {size} sq ft → Price: ${price}k")

is_increasing = True
for i in range(1, len(prices)):
  if prices[i] <= prices[i-1]:
    is_increasing = False

if is_increasing:
    print("\nPositive trend")
```

```python
study_hours = [2, 4, 6, 8]
scores = [55, 65, 75, 85]

total_increase = 0
for i in range(len(study_hours) - 1):
    total_increase += scores[i+1] - scores[i]

avg_increase = total_increase / (len(study_hours) - 1)

predicted_score = scores[-1] + avg_increase

print("10 hours:", predicted_score)
```

```python
scores = [35, 58, 62, 74, 89, 45]
results = []

for score in scores:
    if score < 50:
        label = "Low"
    elif 50 <= score <= 70:
        label = "Medium"
    else:
        label = "High"

    results.append((score, label))

print(results)
```

```python
ages = [18, 20, 22, 24, 26]
heights = [160, 165, 170, 175, 180]

for age, height in zip(ages, heights):
    print(f"Age: {age} → Height: {height} cm")

avg_age = sum(ages) / len(ages)
print(f"\nAverage Age: {avg_age}")

avg_height = sum(heights) / len(heights)
print(f"Average Height: {avg_height} cm")

if heights[-1] > heights[0]:
    print("height appears to increase consistently with age.")
else:
    print("Height does not appear to increase with age.")
```
